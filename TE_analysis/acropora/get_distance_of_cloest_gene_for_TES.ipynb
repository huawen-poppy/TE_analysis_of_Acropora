```python
import pandas as pd
import numpy as np
```

```python
# get the target TEs
targets=pd.read_csv('./data/pure_te_markers_celltype.csv')
targets
```

```python
# remove the unneeded columns in target tes
targets=targets[['cluster','gene']]
targets['gene_id']=targets.gene
targets['gene_id']=targets['gene_id'].replace("#.*","",regex=True)
targets['gene_id']=targets['gene_id'].replace("-","_",regex=True)
targets
```

```python
# read the bed files
bed = pd.read_csv('./get_te_gtf/all_nearest_gene_to_TEs_same_strain.bed',sep='\t',header=None)
bed
```

```python
# filter the bed files based on the targets tes
filter_bed=bed[bed[3].isin(targets.gene_id)]
filter_bed=filter_bed.reset_index(drop=True)
filter_bed=filter_bed.merge(targets,left_on=3,right_on='gene_id')
filter_bed=filter_bed.reset_index(drop=True)
filter_bed=filter_bed.drop_duplicates()
```

```python
filter_bed
```

```python
filter_bed.iloc[10085]
```

```python
all_distance_filter={}
all_distance={}
```

```python
def add_element(dicts,key, value):
    if key not in dicts.keys():
        dicts[key]=[value]
    else:
        dicts[key].append(value)
    return dicts
```

```python
test=filter_bed.copy()
for i in range(filter_bed.shape[0]):
    a=test[1][i]
    b=test[2][i]
    c=test[11][i]
    d=test[12][i]
    if a < b < c < d:
        distance=c-b
        label='far'
    if a < c < b < d:
        distance=b-c
        label='intersect'
    if a < c < d < b:
        distance=0
        label='withinTE'
    if c < a < b < d:
        distance=0
        label='withinGene'
    if c < a < d < b:
        distance=d-a
        label='intersect'
    if c < d < a < b:
        distance=a-d
        label='far'
    output=[distance,label]
    add_element(all_distance,test['cluster'][i]+'='+test['gene'][i]+'='+test[13][i],output)
```

```python
all_distance
```

```python
def remove_duplicate_values(d):
    new_d = {}
    for key, value in d.items():
        new_d[key] = [list(x) for x in set(tuple(x) for x in value)]
    return new_d
```

```python
all_distance_filter=remove_duplicate_values(all_distance)
```

```python
all_distance_filter
```

```python
# get the cloest distance
all_distance_filter_new={}
for key,value in all_distance_filter.items():
    if len(value)==1:
        all_distance_filter_new[key]=value[0]
    else:
        smallest = min(value, key=lambda x: x[1])
        all_distance_filter_new[key]=smallest
```

```python
all_distance_filter_new
```

```python
final_data=pd.DataFrame.from_dict(all_distance_filter_new,orient='index')
final_data.T
```

```python
final_data
```

```python
final_data['full_name']=final_data.index
```

```python
final_data=final_data.reset_index()
```

```python
final_data[['cluster','te_id','gene_id']]=final_data['full_name'].str.split('=',expand=True)
final_data
```

```python
final_data=final_data.rename(columns={0:'distance',1:'label'})
```
